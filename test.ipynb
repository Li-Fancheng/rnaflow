```python
from Bio.PDB import PDBParser

parser = PDBParser()
structure = parser.get_structure("RNA", "ckpt/se3-fm/rna-frameflow/na_sample_0_idx_0_len_66.pdb")

# 查看所有残基的序列号和链
for model in structure:
    for chain in model:
        print(f"Chain: {chain.id}")
        residues = list(chain.get_residues())
        for res in residues:
            print(f"  {res.id[1]:>4} {res.get_resname()}")  # 序列号 + 名称
```

```python
import py3Dmol

# 读取本地 PDB 文件
with open("ckpt/se3-fm/rna-frameflow/na_sample_0_idx_10_len_48.pdb") as f:
    pdb_data = f.read()
if not pdb_data.strip():
    print("❌ PDB 文件为空！")
else:
    print(f"✅ 读取到 {len(pdb_data)} 字符")
view = py3Dmol.view(width=600, height=400)
view.addModel(pdb_data, 'pdb')
view.setStyle({}, {'stick': {}, 'sphere': {'radius': 0.2}})
view.zoomTo()
view.show()
```

```python
import nglview as nv

view = nv.show_file("ckpt/se3-fm/rna-frameflow/na_sample_0_idx_0_len_66.pdb")
view.clear()
# 显式告诉 nglview 这是核酸（RNA/DNA）
view.add_cartoon(color='rainbow', type='nucleic')
# 方案1：只画主链（更容错）
view.add_backbone(color='blue', width=2.0)
view.add_ball_and_stick()
view
```
